```python
import pandas as pd

# Define the amenity groups as categories
amenity_groups = [
    # Education
    ['school', 'university', 'college', 'kindergarten', 'music_school', 'language_school', 'prep_school', 'trade_school', 'dojo', 'driving_school', 'tutor', 'training', 'research_institute'],
    
    # Health and Emergency Services
    ['hospital', 'clinic', 'pharmacy', 'doctors', 'dentist', 'chiropractor', 'nursing_home', 'veterinary', 'first_aid', 'audiologist', 'nursing_room', 'healthcare'],
    
    # Recreation and Outdoors
    ['park', 'fountain', 'bbq', 'garden', 'biergarten', 'cinema', 'theatre', 'nightclub', 'casino', 'karaoke_box', 'spa', 'playground', 'planetarium', 'stage', 'studio', 'sports_centre', 'bench', 'gazebo', 'photo_booth', 'monument', 'hunting_stand', 'binoculars'],
    
    # Food and Beverage
    ['restaurant', 'cafe', 'fast_food', 'bar', 'pub', 'food_court', 'ice_cream', 'canteen', 'internet_cafe', 'hookah_lounge', 'canteen', 'music_venue', 'music_venue;cafe'],
    
    # Public Services
    ['fire_station', 'police', 'post_office', 'townhall', 'courthouse', 'library', 'community_centre', 'registry_office', 'social_facility', 'rescue_station', 'mountain_rescue', 'public_building', 'polling_station', 'ranger_station', 'social_centre', 'public_bath', 'animal_shelter', 'animal_boarding', 'reception_desk', 'mailroom'],
    
    # Commercial Services
    ['bank', 'coworking_space', 'money_transfer', 'bureau_de_change', 'marketplace', 'atm', 'store', 'mall', 'convenience', 'department_store', 'parcel_locker', 'payment_terminal', 'equipment_rental_agency', 'laundramat', 'laundry', 'printer', 'financial_advice'],
    
    # Transportation
    ['bus_station', 'taxi', 'parking', 'parking_entrance', 'parking_space', 'bicycle_parking', 'motorcycle_parking', 'car_rental', 'bicycle_rental', 'truck_rental', 'ferry_terminal', 'rv parking', 'car_pooling', 'trolley_bay', 'boat_rental', 'motorcycle_rental', 'bicycle_repair_station', 'ski_rental'],
    
    # Utilities and Infrastructure
    ['fuel', 'waste_disposal', 'waste_transfer_station', 'recycling', 'watering_place', 'compressed_air', 'charging_station', 'sanitary_dump_station', 'device_charging_station', 'drinking_water', 'loading_dock', 'well', 'storage_box', 'lockers', 'vacuum_cleaner', 'waste_basket', 'waste_basket;vending_machine', 'locker'],
    
    # Religious and Ceremonial
    ['place_of_worship', 'cemetery', 'crematorium', 'monastery', 'grave_yard', 'mortuary', 'funeral_home'],
    
    # Accommodation and Living
    ['apartments', 'nursing_home', 'animal_boarding', 'animal_shelter', 'lodging', 'trailer_park', 'vacation_rental', 'retirement_home'],
    
    # Entertainment and Leisure
    ['arts_centre', 'events_venue', 'conference_centre', 'biergarten', 'stripclub', 'gambling', 'karaoke_box', 'casino', 'studio', 'music_venue', 'garden_centre'],
    
    # Miscellaneous
    ['telephone', 'public_bookcase', 'post_box', 'post_depot', 'letter_box', 'security_control', 'clock', 'photo_booth', 'check_in', 'check-in', 'chair', 'lounger', 'construction', 'feeding_place', 'slide', 'indoor_mausoleum', 'outdoor_mausoleum']
]

# Read the CSV file
df = pd.read_csv('/Users/hadisre/Desktop/PhD - 1 Year/Code/jupyter/polite_env/dataset/england_amenities_with_location.csv')

# Define a function to categorize the amenity
def categorize_amenity(amenity):
    for category, group in zip(
        ['Education', 'Health and Emergency Services', 'Recreation and Outdoors', 'Food and Beverage', 
         'Public Services', 'Commercial Services', 'Transportation', 'Utilities and Infrastructure', 
         'Religious and Ceremonial', 'Accommodation and Living', 'Entertainment and Leisure', 'Miscellaneous'], 
        amenity_groups):
        
        if amenity in group:
            return category
    return 'Uncategorized'

# Apply the categorization function to the 'amenity' column
df['category'] = df['amenity'].apply(categorize_amenity)

# Select only the relevant columns: 'amenity', 'lat', 'lon', 'category'
df_filtered = df[['amenity', 'lat', 'lon', 'category']]

# Save the filtered dataframe to a new CSV file
df_filtered.to_csv('categorized_amenities_filtered.csv', index=False)

# Print the first few rows of the updated dataframe to confirm
print(df_filtered.head())
```
